```python
# import the pandas library
import pandas as pd
import os
import pyodbc
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# import zipfile library (we will use this to extract the file downloaded from Kaggle)
import zipfile

# import kaggle library (we will use this to download the dataset programatically from Kaggle)
import requests
from kaggle.api.kaggle_api_extended import KaggleApi
api = KaggleApi()
api.authenticate()
```

```python
# Difine address of the kaggle dataset
dataset = 'hmavrodiev/london-bike-sharing-dataset'

# Download the dataset file to a temporary location
api.dataset_download_files(dataset, path='./', unzip=False)

# Extract the dataset file from the downloaded ZIP file
with zipfile.ZipFile('london-bike-sharing-dataset.zip', 'r') as z:
    file_name = z.namelist()[0]
    with z.open(file_name) as f:
        # Read directly into a pandas DataFrame
        bikes = pd.read_csv(f)

# Optionally, remove the downloaded ZIP file
os.remove('london-bike-sharing-dataset.zip')

# Convert timestamp column in DataFrame to datetime

# Display the DataFrame
print(bikes.head())
```

```python
# specifying the column names that I want to use
new_cols_dict ={
    'timestamp':'timestamp',
    'cnt':'counts', 
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

# Renaming the columns to the specified column names
bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
# changing the humidity values to percentage (i.e. a value between 0 and 1)
bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
# Retrieve connection parameters from environment variables
server = os.getenv('SERVER')
database = os.getenv('DATABASE')
username = os.getenv('USER')
password = os.getenv('PASSWORD')
table_name = os.getenv('TABLE')

connection_string = f'DRIVER={{ODBC Driver 17 for SQL Server}};SERVER={server};DATABASE={database};UID={username};PWD={password}'
conn = pyodbc.connect(connection_string)
cursor = conn.cursor()
print(table_name)
```

```python
# Insert and check data into SQL Server table
# Get column names from DataFrame
columns = bikes.columns.tolist()

# Create the placeholders for the SQL INSERT statement
placeholders = ', '.join(['?' for _ in columns])
column_names = ', '.join(columns)

# Create the SQL INSERT statement
insert_sql = f"INSERT INTO {table_name} ({column_names}) VALUES ({placeholders})"

timestamp_column = 'timestamp'  # Replace with your actual timestamp column name
def get_latest_timestamp(cursor, table_name, timestamp_column):
    query = f"SELECT MAX({timestamp_column}) FROM {table_name}"
    cursor.execute(query)
    latest_timestamp = cursor.fetchone()[0]
    if latest_timestamp and isinstance(latest_timestamp, str):
        latest_timestamp = pd.to_datetime(latest_timestamp)
    return latest_timestamp

# Get the latest timestamp from the table
latest_timestamp = get_latest_timestamp(cursor, table_name, timestamp_column)
print(f"Latest timestamp: {latest_timestamp}")

# Insert data into SQL Server table
for index, row in bikes.iterrows():
    data_timestamp = pd.to_datetime(row[timestamp_column], errors='coerce')

    if latest_timestamp is None or (data_timestamp > latest_timestamp):
        cursor.execute(insert_sql, tuple(row))
        print(f"Inserted row {index}.")
    else:
        print(f".")

# Commit the transaction
conn.commit()

cursor.close()
conn.close()

print("Finish")
```
